```python
from pybaseball import batting_stats
import pandas as pd

df = batting_stats(1900)
df.head()
```

```python
df = df[df['PA'] >= 200].reset_index(drop=True)
df_2023 = batting_stats(2023)
df_2024 = batting_stats(2024)

merged = pd.merge(df_2023, df_2023[['Name', 'OPS']], on='Name', suffixes=('', '_target'))
```

```python
features = [
    'G', 'AB', 'H', 'HR', 'BB', 'SO', 'SB', '2B', '3B', 'OPS', 'OBP', 'SLG', 'wOBA', 
    'AVG', 'OBP', 'SLG', 'OPS', 'ISO', 'BABIP', 'BB%', 'K%', 'BB/K', 'HR/FB', 'wOBA', 
    'wRC', 'wRC+', 'WAR', 'LD%', 'GB%', 'FB%', 'Pull%', 'Cent%', 'Oppo%', 'Soft%', 
    'Med%', 'Hard%', 'EV', 'LA', 'Barrel%', 'HardHit%', 'O-Swing%', 'Z-Swing%', 
    'Swing%', 'O-Contact%', 'Z-Contact%', 'Contact%', 'SwStr%', 'Zone%', 'F-Strike%'
    ]

merged_data = merged[features + ['OPS_target']].dropna()
merged_data.to_csv('../data/processed_player_data.csv', index=False)
```
